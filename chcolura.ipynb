```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\pedro\\Downloads\\material+curso+-+Time+Series\\material curso - Time Series\\chocolura.csv')
df.head()
```

```python
print("Exibindo a quantidade de linhas e de colunas: ", df.shape)
```

```python
df.dtypes
```

```python
df['mes'] = pd.to_datetime(df['mes'])
df.dtypes
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
df['aumento'] = df['vendas'].diff()
df.head()
```

```python
df['aceleracao'] = df['aumento'].diff()
df.head()
```

Um otimo exemplo de sazonalidade, onde entre abril e em junho dos dois anos se mantiveram com vendas parecidas

```python
plt.figure(figsize=(16, 12))
ax = plt.subplot(3,1,1)
ax.set_title('Análise de vendas da chocolura de 2017 e 2018', fontsize=18, loc='left')
sns.lineplot(x='mes', y='vendas', data=df)
plt.subplot(3,1,2)
sns.lineplot(x='mes', y='aumento', data=df)
plt.subplot(3,1,3)
sns.lineplot(x='mes', y='aceleracao', data=df)
ax=ax
```

```python
from pandas.plotting import autocorrelation_plot

ax = plt.figure(figsize=(12, 6))
ax.suptitle('Correlação das vendas', fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(df['vendas'])
ax = ax
```
