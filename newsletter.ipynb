```python
df['mes'] = pd.to_datetime(df['mes'])
df.dtypes
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
df['aumento'] = df['assinantes'].diff()
df.head()
```

```python
df['aceleracao'] = df['aumento'].diff()
df.head()
```

```python
plt.figure(figsize=(16, 12))
ax = plt.subplot(3,1,1)
ax.set_title('Análise de assinantes Newsletter', fontsize=18, loc='left')
sns.lineplot(x='mes', y='assinantes', data=df)
plt.subplot(3,1,2)
sns.lineplot(x='mes', y='aumento', data=df)
plt.subplot(3,1,3)
sns.lineplot(x='mes', y='aceleracao', data=df)
ax=ax
```
